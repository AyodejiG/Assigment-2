```python
import selenium
import pandas as pd 
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
#connecting to the webdriver
driver=webdriver.Chrome(r"chromedriver.exe")
```

```python
#Opening the Naukri page on automated chrome browser
driver.get("https://www.shine.com/")
```

```python
# entering designation and location  as required in the assignment
designation=driver.find_element(By.CLASS_NAME, "form-control")
designation.send_keys('Data Analyst')
```

```python
location=driver.find_element(By.XPATH,"/html/body/div[1]/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input")
location.send_keys('Bangalore')
```

```python
search=driver.find_element(By.CLASS_NAME,"searchForm_btnWrap_advance__VYBHN")
search.click()
```

```python
job_title=[]
job_location=[]
company_name=[]
experience_required=[]
```

```python
#scraping Job title from the page
title_tags=driver.find_elements(By.XPATH,'//h2[@itemprop="name"]')
for i in title_tags[0:10]:
    title=i.text
    job_title.append(title)
    

# scraping for job location
location_tags=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in location_tags[0:10]:
    location=i.text
    job_location.append(location)

#scraping for the company name 
company_tags=driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]')
for i in company_tags[0:10]:
    company=i.text
    company_name.append(company)

#scraping for job experience 
experience_tags=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in experience_tags[0:10]:
    experience=i.text
    experience_required.append(experience)
```

```python
print(len(job_title),len(job_location),len (company_name), len(experience_required))
```

```python
import pandas as pd 
df=pd.DataFrame({'Title': job_title,'Location': job_location, 'Company Name': company_name, 'Experience':experience_required})
df
```

```python
# Q10: Write a python program to display list of 50 Most expensive cars in the world (i.e.Car name and Price) from https://www.motor1.com/


import selenium
import pandas as pd 
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
#connecting to the webdriver
driver=webdriver.Chrome(r"chromedriver.exe")
```

```python
#Opening the Naukri page on automated chrome browser
driver.get("https://www.motor1.com/")
```

```python
# entering designation 50 most expensive cars as required in the assignment
designation=driver.find_element(By.CLASS_NAME,"m1-search-form-text")
designation.send_keys('50 most expensive car')




```

```python
cars=driver.find_element(By.XPATH,"/html/body/div[3]/div[2]/div/div/div[3]/div/div/div/form/button[1]")
cars.click()
```

```python
cars_name=[]
car_price=[]
```

```python
#scraping for car name and price
cars_tags=driver.find_elements(By.XPATH,'//h3[@class="subheader"]')
for i in cars_tags[0:40]:
    cars=i.text
    cars_name.append(cars)
    

# scraping for job location
price_tags=driver.find_elements(By.XPATH,'//a[@data-inline-widget="internal-links"]')
for i in price_tags[0:40]:
    price=i.text
    car_price.append(price)
```

```python
print(len(cars_name),len(car_price))
```

```python
df=pd.DataFrame({'Car Name': cars_name,'Car Price': car_price})
df
```
