```python
# Question 1. Write a python program which searches all the product under a particular product from www.amazon.in. The
product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for
guitars.


import selenium
import pandas as pd 
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
#connecting to the webdriver
driver=webdriver.Chrome(r"chromedriver.exe")
```

```python
#Opening the amazon page on automated chrome browser
driver.get("https://www.amazon.in/")
```

```python
# entering USER INPUT  IN SERACH BOX as required in the assignment
designation=driver.find_element(By.ID, "twotabsearchtextbox")
designation.send_keys('Guitar')
```

```python
submitsearch=driver.find_element(By.ID,"nav-search-submit-button")
submitsearch.click()
```

```python
Available_Guitar=[]
Price_sold=[]
```

```python
Guitar_tags=driver.find_elements (By.XPATH,'//span[@class="a-size-base-plus a-color-base a-text-normal"]')
for i in Guitar_tags[0:50]:
    Guitar=i.text
    Available_Guitar.append(Guitar)
    
Price_tags=driver.find_elements(By.XPATH,'//span[@class="a-price-whole"]')
for i in Price_tags[0:50]:
    Price=i.text
    Price_sold.append(Price)
```

```python
print(len(Available_Guitar),len(Price_sold))
```

```python
df=pd.DataFrame({'Guitar':Available_Guitar,'Price': Price_sold})
df
```

```python
# question 5 Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps. 


import selenium
import pandas as pd 
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
```

```python
#connecting to the webdriver
driver=webdriver.Chrome(r"chromedriver.exe")
```

```python
driver.get("https://www.google.com/maps")
```

```python
city = input("Enter the city name: ")
```

```python
search = driver.find_element(By.CLASS_NAME,"searchboxinput")
search.send_keys(city)
```

```python
entersearch=driver.find_element(By. CLASS_NAME,'mL3xi')
entersearch.click()
```

```python
html = driver.page_source
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
coordinate_element = soup.select_one("meta[itemprop=image]")
```

```python
if coordinate_element:
    # Extract the coordinates from the content attribute
    content = coordinate_element["content"]
    coordinates = content.split(";"),(",")
```

```python
latitude = coordinates[0],()
longitude = coordinates[1],()
```

```python
print("Latitude:", latitude)
print("Longitude:", longitude)


    
       
```

```python
df=pd.DataFrame({'Latitude': (latitude), 'Longitude' : (longitude)})
df
```

```python
# question 9. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in
“London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall
reviews, privates from price, dorms from price, facilities and property description. 
```

```python
#connecting to the webdriver
driver=webdriver.Chrome(r"chromedriver.exe")
```

```python
#Opening the google page on automated chrome browser
driver.get("https://www.hostelworld.com/")
```

```python
# entering USER INPUT  IN SERACH BOX as required in the assignment
designation=driver.find_element(By.ID,"search-input-field")
designation.send_keys('LONDON')
```

```python
search=driver.find_element(By.CLASS_NAME,"search-button")
search.click()
```

```python
hostelName=[]
distance_town=[]
ratings=[]
privateprice=[]
dormprice=[]
description=[]

```

```python
hostel_tags=driver.find_elements(By.XPATH,'//div[@class="property-name"]')
for i in hostel_tags[0:25]:
    hostel=i.text
    hostelName.append(hostel)
    
    
distance_tags=driver.find_elements(By.XPATH,'//span[@class="distance-description"]')
for i in distance_tags[0:25]:
    distance=i.text
    distance_town.append(distance)
    
    
rating_tags=driver.find_elements(By.XPATH,'//div[@class="keyword"]')
for i in rating_tags[0:25]:
    rating=i.text
    ratings.append(rating)
    
    
private_tags=driver.find_elements(By.XPATH,'//div[@class="accommodation-label"]')
for i in private_tags[0:25]:
    private=i.text
    privateprice.append(private)
    
    
dom_tags=driver.find_elements(By.XPATH,'//div[@class="accommodation-label"]')
for i in dom_tags[0:25]:
    dom=i.text
    dormprice.append(dom)
```

```python
print(len(hostelName),len(distance_town),len(ratings),len(privateprice),len(dormprice))
```

```python
df=pd.DataFrame({'HOSTEL NAME': (hostelName), 'DISTANCE TO TOWN' : (distance_town), 'RATING': (ratings), 'PRIVATE PRICE': (privateprice), 'DORM PRICE': (dormprice)})
df
```
